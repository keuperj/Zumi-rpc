```python
import rpyc
import numpy as np

#pickling of return objects must be turned on 
rpyc.core.protocol.DEFAULT_CONFIG['allow_pickle'] = True

ZUMI_IP = '192.168.10.1' #default IP  if connected to umi Wifi
ZUMI_PORT = 9004

#conn = rpyc.ssl_connect(ZUMI_IP, ZUMI_PORT, keyfile="client.key", certfile="client.crt")

conn = rpyc.connect(ZUMI_IP, ZUMI_PORT)
zumi = conn.root
```

```python
# ping the zumi
ping = zumi.ping()
print(ping)
```

```python
#move forward
zumi.forward()
```

```python
zumi.go_reverse(1,0)
```

```python
zumi.hard_brake()
```

```python
#get image from cam
# need to cat to LOCAL numpy array, otherwise the result stays remote!
frame = np.array(zumi.get_picture())
```
